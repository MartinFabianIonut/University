```python
reset()
w0=var('w0')
w=var('w')

assume(w0>0)
t=var('t')
x0=var('x0')
v0=var('v0')
x=function('x')(t)

F=function('F')(t)
F0=var('F0')
ecF=F0*cos(w*t)

eqd=diff(x,t,2)+w0^2*x==ecF

assume(w0!=w)
show(desolve(eqd,[x,t]))
```

```python
sol(t,w0,w,F0,x0,v0)=desolve(eqd,[x,t],ics=[0,x0,v0])
show(sol(t,5,5.5,2,0,0))
plot(sol(t,5,5.5,2,0,0),t,0,30)
```
